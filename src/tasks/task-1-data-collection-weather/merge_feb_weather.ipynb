```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/Feb2022_train.csv')
```

```python
df.info()
```

```python
w = pd.read_csv('data/Berlin_weather.csv', parse_dates=['time'])
```

```python
w['date'] = w['time'].dt.date
```

```python
import unicodedata
def strip_accents(s):
    nfkd_form = unicodedata.normalize('NFKD', s)
    only_ascii = nfkd_form.encode('ASCII', 'ignore')
    return only_ascii.decode()
```

```python
df['city_code'] = df['DESTINATION'].str.upper().str.slice(0,4)
w['city_code'] = w['city'].str.upper().str.slice(0,4).apply(strip_accents)
```

```python
wd = w.set_index('time').resample('d').first()
```

```python
wd = w.groupby([
    pd.Grouper(freq='d', key='time'),
    'city_code'
]).first().reset_index()
```

```python
wd
```

```python
df['id'] = df['DATDEP'].astype(str)+df['city_code']
wd['id'] = wd['time'].dt.date.astype(str)+wd['city_code']
```

```python
merged = pd.merge(df, wd, how='left', left_on=['id'], right_on=['id'])  
```
